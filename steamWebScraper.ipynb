```python
rev1 = []
rev2 = []
tags = []
precos = []
titulos = []
```

```python
i = 0
j = 0
```

```python
while(i < len(jogos3)):
    
    url = jogos3[i][:-21]
    driver.get(url)
    time.sleep(3)
    
    if(len(driver.find_elements(By.ID, 'ageYear')) != 0):
        
        select = Select(driver.find_element(By.ID, 'ageYear'))
        select.select_by_value('1990')
        time.sleep(2)
        
        b = driver.find_element(By.ID,'view_product_page_btn')
        b.click()
        time.sleep(3)
        
    tags_t = driver.find_elements(By.CSS_SELECTOR, 'a.app_tag')
    title = driver.find_elements(By.CSS_SELECTOR, 'div#appHubAppName.apphub_AppName')
    preco = driver.find_elements(By.CSS_SELECTOR, 'div.discount_original_price')
        
    element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div.user_reviews_filter_menu')))
    clickables = driver.find_elements(By.CSS_SELECTOR, 'div.user_reviews_filter_menu')
    time.sleep(2)
    clickables[0].click()
    
    time.sleep(2)
    
    element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'span.user_reviews_count')))
    reviews = driver.find_elements(By.CSS_SELECTOR, 'span.user_reviews_count')
    
    j = j + 1
    
    if(j > 2):
        jogos3.remove(jogos3[i])
        i = i + 1
        j = 0
        
    texto = reviews[0].text.replace('(', '').replace(',', '').replace(')','')
    
    if(texto != ''):
        if(len(tags_t) > 2):
            rev1.append(int(reviews[0].text.replace('(', '').replace(',', '').replace(')','')))
            rev2.append(int(reviews[1].text.replace('(', '').replace(',', '').replace(')','')))
            tags.append((tags_t[0].text, tags_t[1].text, tags_t[2].text))
            precos.append(preco[0].text)
            titulos.append(title[0].text)
            time.sleep(2)
            j = 0
        texto = ''
        i = i + 1
```

```python
jogos = list(zip(titulos, precos, rev1, rev2, tags))
```

```python
df = pd.DataFrame(jogos, columns =['Titulo', 'Preco', 'ReviewTotal', 'ReviewPositivo', 'Tags'])
```
